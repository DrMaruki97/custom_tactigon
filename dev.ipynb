```python
import pandas as pd
```

```python
csv = pd.read_csv(r'C:\Users\LucaAsara\OneDrive - ITS Angelo Rizzoli\Desktop\ITS 23-25\UFS\UFS 15 big data e IOT\custom_tactigon\movement_data\tact_base_move.csv')
```

```python
divided_dfs = []

for i in csv["label"].unique():
    df = csv[csv["label"] == i]
    divided_dfs.append(df)
```

```python
len(divided_dfs)
```

```python

for x in range(len(divided_dfs)):
    divided_dfs[x] = divided_dfs[x][:1500]
```

```python
lista = []

for i in divided_dfs:
    for j in range(0,len(i),100):
        lista.append(i[j:j+100]) 
        
print(len(lista))
```

```python
for x in lista:
    x.drop("Unnamed: 0", axis = 1, inplace=True)
```

```python
labels = []

for x in lista:
    label = x['label'].unique()[0]
    x.drop("label", axis=1, inplace=True)
    data = x.stack().tolist()
    data.append(label)
    labels.append(pd.DataFrame(data))
```

```python
type(labels)
```

```python
for i,x in enumerate(labels):
    x = x.transpose()
    labels[i] = x
```

```python
type(labels)
```

```python
finaldf = pd.concat(labels)
```
